```python
from PIL import Image
import os
import sys
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = Image.open('D:\\Dokumente\\03_RCI\\practical\\Folder_structure\\Registration\\images\\sub-0030_slice_015.png')
	
#img = (img - np.min(img)) / (np.max(img) - np.min(img)) * 2 - 1
# NORMALIZE IMAGE TO 0, 1
#img = (img - np.min(img)) / (np.max(img) - np.min(img))
# normalize image to 0, 255
img = ((img - np.min(img)) / (np.max(img) - np.min(img)) * 255).astype(np.uint8)
print(np.min(img), np.max(img))
img_edge = cv2.Canny(img.astype(np.uint8), 200, 300)
img_edge = np.array(img_edge)


    
print(img_edge.shape)
print(img.shape)

if len(img.shape) == 2 or img.shape[2] == 1:
    img = cv2.cvtColor(img, cv2.COLOR_GRAY2BGR)

#overlayed = cv2.addWeighted(img, 0.6, cv2.cvtColor(img_edge, cv2.COLOR_GRAY2BGR), 0.4, 0)
overlayed = img.copy()
overlayed[img_edge > 0] = [255, 0, 0]  # Red color for edges
```

```python
# Display the images
plt.imshow(overlayed)
plt.axis('off')
plt.show()

```
